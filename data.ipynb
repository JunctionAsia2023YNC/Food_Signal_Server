```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
food = pd.read_csv(r'data.csv', encoding='cp949')
```

```python
food.head()
```

```python
r=food.columns
r
```

```python
food_dropped=food.drop(columns=[r[4]])
food_dropped.columns
```

```python
food_dropped["대표메뉴"].unique()
```

```python
keys = ["고기", "국수", "회", "초밥", "짜장면", "돈가스", ("자장면","짜장면"), ("닭", "닭고기"), "샤브샤브", "짬뽕", "갈비", "곰탕", ("중", "중식"),
        "낙지", ("돈까스", "돈가스"), "광어", ("생 갈 비", "생갈비"), ("부페", "뷔페")
        ]
for keyword in keys+sorted(food_dropped['대표메뉴'].unique()):
    if type(keyword)==tuple and len(keyword)==2:
        key=keyword[1]
        keyword=keyword[0]
    else: key=keyword
    food_dropped.loc[food_dropped['대표메뉴'].str.contains(keyword), '대표메뉴'] = key
print(len(food_dropped['대표메뉴'].unique()))
print(sorted(food_dropped['대표메뉴'].unique()))
```
